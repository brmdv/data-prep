## Coordinates and addresses
At first sight, most dtypes look to be correct. De features *the_geom* and (*longitude*, *latitude*) seem to be interchangeable, so we look into that.

```python
fake_long=df["the_geom"].apply(lambda x: np.float64(x[7:-1].split(" ")[0]))
fake_lat=df["the_geom"].apply(lambda x: np.float64(x[7:-1].split(" ")[1]))
print(f"Mean difference between longitude: {np.mean(np.abs(df['longitude']-fake_long))}")
print(f"Mean difference between  latitude: {np.mean(np.abs(df['latitude']-fake_lat))}")
```

so very small, although the _the_geom_ is more precisely saved in the csv. So just to be sure, we replace the longitude and latitude with that. (The difference could just come from floating-point precision mismatces, but I don't want to go down that path.)

```python
df["longitude"]=fake_long
df["latitude"]=fake_lat
df.drop("the_geom", axis=1,inplace=True)
df.head()
```

The columns *x_sp* and *y_sp* specify the location in the plane of New York State. So this is probably also double information that could be recalculated if really needed. I choose to drop it. 

Furthermore, every trees is in New York State, so we don't need to repeat that for every row.

```python
df.drop(["state", "x_sp", "y_sp"], axis=1, inplace=True)
```

## Consolidate text values
We look at some textual values now, to see if we can consolidate them.

```python
df.loc[:,"status":"brnch_othe"].head()
```

All the yes/no vfeatures can easily be changed to 0/1.

```python
df.loc[:,"root_stone":"brnch_othe"].replace(to_replace=['No', 'Yes'], value=[0, 1], inplace=True)
```

```python
df.loc[:,"root_stone":"brnch_othe"].head()
```

For the other terxt values, we take a look at how many unique ones there are.

```python
for colname in df.loc[:,"curb_loc":"problems"].columns:
    uniqvals=pd.unique(df[colname])
    print(f"{colname:>10}: {len(uniqvals)}\n{uniqvals}\n")
```

Let's go over them one by one.

### Species name
The tree species is provides in Latin as well as in common English. There should be a one to onerelation between them, so it would be better to save one of them in another table. For this simple case, I will just drop the Latin names.

```python
df.drop(["spc_latin"],axis=1, inplace=True)
```

### curb_loc, damage
These are binary choices, so we change them.

```python
df["curb_loc"].replace(["OnCurb", "OffsetCurb"], [1,0], inplace=True)
df["sidewalk"].replace(["Damage", "NoDamage"], [1,0], inplace=True)
df.rename(columns={"curb_loc":"is_OnCurb", "sidewalk":"sidewalk_damaged"}, inplace=True)
```

### status
Is "alive", "dead" or "stump", so same information can be condensed to "is_dead" and "is_stump".

```python
status=pd.get_dummies(df["status"], drop_first=True)
```

```python
df["is_dead"]=status.iloc[:,0]
df["is_stump"]=status.iloc[:,1]
df.drop("status", axis=1, inplace=True)
```

### health
Can be made numeric


```python
df["health"].replace(["Good", "Fair", "Poor"], [2,1,0], inplace=True)
```

### steward

This is a string representing a range of numbers, so can be represented by the lower limit.

```python
['None' '1or2' '3or4' nan '4orMore']
df["steward"].replace(['None' '1or2' '3or4' nan '4orMore'], [2,1,0], inplace=True)
```
