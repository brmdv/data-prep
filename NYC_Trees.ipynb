```python
import pandas as pd
import numpy as np
```

Import the dataset.

```python
df=pd.read_csv("trees_1000.csv", parse_dates=[0])
df.head()
```

```python
df.info()
```

Set index.

```python
df.set_index("tree_id", inplace=True)
df.head()
```

At first sight, most dtypes look to be correct. De features *the_geom* and (*longitude*, *latitude*) seem to be interchangeable, so we look into that.

```python
fake_long=df["the_geom"].apply(lambda x: np.float64(x[7:-1].split(" ")[0]))
fake_lat=df["the_geom"].apply(lambda x: np.float64(x[7:-1].split(" ")[1]))
print(f"Mean difference between longitude: {np.mean(np.abs(df['longitude']-fake_long))}")
print(f"Mean difference between  latitude: {np.mean(np.abs(df['latitude']-fake_lat))}")
```

so very small, although the _the_geom_ is more precisely saved in the csv. So just to be sure, we replace the longitude and latitude with that. (The difference could just come from floating-point precision mismatces, but I don't want to go down that path.)

```python
df["longitude"]=fake_long
df["latitude"]=fake_lat
df.drop("the_geom", axis=1,inplace=True)
df.head()
```
